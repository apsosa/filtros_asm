```python
import numpy as np                   # Para cosas de 치lgebra lineal
import matplotlib.pyplot as plt      # Para gr치ficos
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
import pandas as pd
import math
import random
import matplotlib.lines as mlines
import matplotlib.patches as mpatches
```

```python
#####Obtenemos datos####

datosC_O0 = pd.read_csv("medidas/variar_O/medidas_C_O0_20it.csv",header=0)
datosC_O1 = pd.read_csv("medidas/variar_O/medidas_C_O1_20it.csv",header=0)
datosC_O2 = pd.read_csv("medidas/variar_O/medidas_C_O2_20it.csv",header=0)
datosC_O3 = pd.read_csv("medidas/variar_O/medidas_C_O3_20it.csv",header=0)

datosASM = pd.read_csv("medidas/variar_O/medidas_asm.csv",header=0)

#####De los datos nos quedamos con los clocks####

cant_clocks_O0 = np.array(datosC_O0["cant_clock"])
cant_clocks_O1 = np.array(datosC_O1["cant_clock"])
cant_clocks_O2 = np.array(datosC_O2["cant_clock"])
cant_clocks_O3 = np.array(datosC_O3["cant_clock"])

cant_clocks_asm = np.array(datosASM["cant_clock"])


#####Obtenemos medianas####

mediana_O0 = np.median(cant_clocks_O0)
mediana_O1 = np.median(cant_clocks_O1)
mediana_O2 = np.median(cant_clocks_O2)
mediana_O3 = np.median(cant_clocks_O3)

mediana_asm = np.median(cant_clocks_asm)


####Hacemos dibujito#####

sns.set_style('darkgrid')

fig=plt.figure()

medianas = [mediana_asm, mediana_O0, mediana_O1, mediana_O2, mediana_O3]

labels = ('ASM', 'O0', 'O1', 'O2', 'O3')

y_pos = np.arange(len(labels))

fig.suptitle('Comparaci칩n ASM vs optimizaci칩n O')

plt.bar(y_pos, medianas, align='center', alpha=0.8)

plt.xticks(y_pos, labels)

plt.ylabel('Clocks')


```
